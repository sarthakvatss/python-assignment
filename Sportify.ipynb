```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("Spotify_Youtube.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# Danceability vs. Streams
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Danceability", y="Stream", data=df, size="Stream", sizes=(50, 500), alpha=0.6)
plt.title("Danceability vs. Spotify Streams", fontsize=16)
plt.xlabel("Danceability (0-1)", fontsize=12)
plt.ylabel("Spotify Streams", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Energy vs. Views
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Energy", y="Views", data=df, size="Views", sizes=(50, 500), alpha=0.6, color="orange")
plt.title("Energy vs. YouTube Views", fontsize=16)
plt.xlabel("Energy (0-1)", fontsize=12)
plt.ylabel("YouTube Views", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Valence vs. Likes
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Valence", y="Likes", data=df, size="Likes", sizes=(50, 500), alpha=0.6, color="green")
plt.title("Valence vs. YouTube Likes", fontsize=16)
plt.xlabel("Valence (0-1)", fontsize=12)
plt.ylabel("YouTube Likes", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Optional: Correlation matrix to see relationships numerically
correlation_matrix = df[['Danceability', 'Energy', 'Valence', 'Stream', 'Views', 'Likes']].corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
# Visualize correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix of Song Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
from io import StringIO
df = pd.read_csv(StringIO(data))
sns.set(style="whitegrid")

#  Artists Comparison: Average Danceability, Energy, Valence
artist_avg = df.groupby('Artist')[['Danceability', 'Energy', 'Valence']].mean().reset_index()
plt.figure(figsize=(10, 6))
artist_avg_melted = artist_avg.melt(id_vars=['Artist'], var_name='Feature', value_name='Value')
sns.barplot(x='Feature', y='Value', hue='Artist', data=artist_avg_melted)
plt.title("Average Musical Features: Gorillaz vs. Red Hot Chili Peppers", fontsize=16)
plt.ylabel("Average Value (0-1)", fontsize=12)
plt.xlabel("Feature", fontsize=12)
plt.legend(title="Artist")
plt.tight_layout()
plt.show()
```

```python
# Songs Comparison: Danceability, Energy, Loudness, Tempo
songs_to_compare = df[df['Track'].isin(['Feel Good Inc.', 'Californication'])]
plt.figure(figsize=(12, 6))
songs_melted = songs_to_compare[['Track', 'Danceability', 'Energy', 'Loudness', 'Tempo']].melt(id_vars=['Track'], var_name='Feature', value_name='Value')
sns.barplot(x='Feature', y='Value', hue='Track', data=songs_melted)
plt.title("Feature Comparison: Feel Good Inc. vs. Californication", fontsize=16)
plt.ylabel("Value", fontsize=12)
plt.xlabel("Feature", fontsize=12)
plt.legend(title="Song")
plt.tight_layout()
plt.show()
```

```python
# Popularity Metrics: Streams, Views, Likes
popularity = df[['Artist', 'Track', 'Stream', 'Views', 'Likes']]
plt.figure(figsize=(12, 6))
popularity_melted = popularity.melt(id_vars=['Artist', 'Track'], var_name='Metric', value_name='Count')
sns.barplot(x='Track', y='Count', hue='Metric', data=popularity_melted)
plt.title("Popularity Metrics: Streams, Views, Likes", fontsize=16)
plt.ylabel("Count", fontsize=12)
plt.xlabel("Track", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title="Metric")
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("Spotify_Youtube.csv")
gorillaz_demon_days = df[(df['Artist'] == 'Gorillaz') & (df['Album'] == 'Demon Days')]

# Summary of Musical Features
musical_features = gorillaz_demon_days[['Danceability', 'Energy', 'Valence', 'Loudness', 'Tempo']].mean()
print("Musical Features Summary for Gorillaz - Demon Days:")
print(musical_features)
```

```python
#Popularity Metrics
popularity_metrics = gorillaz_demon_days[['Stream', 'Views', 'Likes', 'Comments']].sum()
print("\nPopularity Metrics Summary for Gorillaz - Demon Days:")
print(popularity_metrics)
```

```python
#Song Details
print("\nSong Details for Gorillaz - Demon Days:")
print(gorillaz_demon_days[['Track', 'Duration_ms', 'Danceability', 'Energy', 'Valence', 'Stream', 'Views']])
```

```python
# Visualization 1: Bar plot of Musical Features
plt.figure(figsize=(10, 6))
musical_features_df = pd.DataFrame(musical_features, columns=['Value']).reset_index()
musical_features_df.columns = ['Feature', 'Average Value']
sns.barplot(x='Feature', y='Average Value', data=musical_features_df, palette='Blues_d')
plt.title("Average Musical Features for Gorillaz - Demon Days", fontsize=16)
plt.ylabel("Average Value", fontsize=12)
plt.xlabel("Feature", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Visualization 2: Bar plot of Popularity Metrics
plt.figure(figsize=(10, 6))
popularity_df = pd.DataFrame(popularity_metrics, columns=['Count']).reset_index()
popularity_df.columns = ['Metric', 'Total Count']
sns.barplot(x='Metric', y='Total Count', data=popularity_df, palette='Greens_d')
plt.title("Total Popularity Metrics for Gorillaz - Demon Days", fontsize=16)
plt.ylabel("Total Count", fontsize=12)
plt.xlabel("Metric", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Visualization 3: Bar plot of Streams per Song
plt.figure(figsize=(12, 6))
sns.barplot(x='Track', y='Stream', data=gorillaz_demon_days, palette='Purples_d')
plt.title("Spotify Streams per Song in Gorillaz - Demon Days", fontsize=16)
plt.ylabel("Spotify Streams", fontsize=12)
plt.xlabel("Track", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
